Interesting questions:

We have two people, Alice and Betty. After graduating in YEAR, Alice does a PhD and Betty goes straight into a graduate job.

We make two few assumptions that may be very generous in reality:
A PhD takes Alice exactly 3 years to complete
Alice is able to start earning a post-graduate salary immediately after finishing her PhD
Neither Alice nor Betty contribute to a pension or invest any of their income: they spend it all and so have no secondary sources of income
Neither Alice nor Betty get a promotion at work; when they enter the workplace Betty stays on a graduate salary forever, and Alice stays on a Graduate salary forever

How long does it take Alice to catch up with Betty in terms of lifetime earnings?

```python
yeary = 2014
```

```python
df = df.pivot(index='year', columns='income_type', values='income').reset_index()
```

```python
df.index = df['year']
```

```python
df
```

```python
# the sum of Graduate from year 2019 to 2022
df_filtered = df[(df.index.year >= yeary) & (df.index.year < yeary+3)]
df_filtered = df_filtered.copy()
df_filtered
```

```python
# Earnings loss
df_filtered['Earnings Loss'] = df_filtered['Stipend'] - df_filtered['Graduate']

# sum of Graduate from year 2019 to 2022
loss = df_filtered[['Earnings Loss']].sum()[0]

print(loss)
```

```python
df_filtered
```

From year onwards, Alice is earning more and Betty is earning less, 

```python
df_filtered = df[(df.index.year >= yeary+3)]
df_filtered = df_filtered.copy()
```

```python
df_filtered["running_gain"] = df_filtered["Postgraduate"] - df_filtered["Graduate"]
df_filtered["cumulative_gain"] = df_filtered["running_gain"].cumsum()
df_filtered
```

```python
break_even_year = df_filtered[df_filtered["cumulative_gain"] >= -loss].index[0]
print(f"Break even year: {break_even_year}")

# the number of years betwen the start of the df and this break even year
years = break_even_year.year - yeary

print(f"it takes {years} years to break even from the start of doing a phd")
```
